1. Decimal a binario

```python
def decimal_a_binario(numero):
    if numero == 0:
        return "0"
    
    binario = ""
    
    while numero > 0:
        residuo = numero % 2
        binario = str(residuo) + binario
        numero = numero // 2
    
    return binario


print(decimal_a_binario(10))
```

2. Número primo

```python
def es_primo(n):
    if n <= 1:
        return False
    
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    
    return True


print(es_primo(7))
print(es_primo(10)) 
```

3. Desviación estándar (sin funciones built-in)

```python
def desviacion_estandar(lista):
    n = len(lista)
    
    # media
    suma = 0
    for num in lista:
        suma += num
    media = suma / n
    
    ## varianza
    suma_varianza = 0
    for num in lista:
        suma_varianza += (num - media) ** 2
    
    varianza = suma_varianza / n
    
    ## desviación
    desviacion = varianza ** 0.5
    
    return desviacion


print(desviacion_estandar([1, 2, 3, 4, 5]))
```

4. Ordenamiento burbuja

```python
def burbuja(arreglo):
    n = len(arreglo)
    nuevo = arreglo.copy()
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if nuevo[j] > nuevo[j + 1]:
                # intercambiar
                nuevo[j], nuevo[j + 1] = nuevo[j + 1], nuevo[j]
    
    return nuevo

def main():
    datos = [5, 2, 9, 1, 5, 6]
    ordenado = burbuja(datos)
    print("Original:", datos)
    print("Ordenado:", ordenado)

main()
```

5. Tupla → String

```python
def tupla_a_string(tupla):
    resultado = ""
    
    for elemento in tupla:
        resultado += str(elemento)
    
    return resultado

# Prueba
print(tupla_a_string((10, 20, 40, 5, 70)))
```

6. Eliminar tuplas vacías

```python
def eliminar_vacias(lista):
    nueva_lista = []
    
    for tupla in lista:
        if tupla:  # si no está vacía
            nueva_lista.append(tupla)
    
    return nueva_lista

# Prueba
lista = [(), (), ('X',), ('a','b'), ('a','b','c'), ('d',)]
print(eliminar_vacias(lista))
```

7. Promedio de tuplas

```python
def promedio_tuplas(tuplas):
    resultado = []
    
    for t in tuplas:
        suma = 0
        for num in t:
            suma += num
        promedio = suma / len(t)
        resultado.append(promedio)
    
    return tuple(resultado)


entrada = ((10,10,10,12),(30,45,56,45),(81,80,39,32))
print(promedio_tuplas(entrada))
```

8. Clase Libro (POO)

```python
class Libro:
    
    def __init__(self, titulo, autor, paginas):
        self.titulo = titulo
        self.autor = autor
        self.paginas = paginas
        self.leido = False
    
    def marcar_leido(self):
        self.leido = True
    
    def descripcion(self):
        estado = "Leído" if self.leido else "No leído"
        return f"Título: {self.titulo}, Autor: {self.autor}, Páginas: {self.paginas}, Estado: {estado}"



libro1 = Libro("Python Básico", "Juan Pérez", 200)
print(libro1.descripcion())

libro1.marcar_leido()
print(libro1.descripcion())
```

```python
class Circulo:
    
    def __init__(self, radio):
        self.radio = radio
    
    def area(self):
        return 3.1416 * (self.radio ** 2)
    
    def circunferencia(self):
        return 2 * 3.1416 * self.radio



c = Circulo(5)
print("Área:", c.area())
print("Circunferencia:", c.circunferencia())
```
